## Data Visualization With Matplotlib

#### Matplotlib is a powerful plotting library for python that enables the creation of static, animated, and interactive visualizations. It is widely used for data visualizations.

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

## Create a line plot

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Line Plot')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

## Create a customized line plot

plt.plot(x, y, color='red', linestyle='--', marker='o', linewidth=2, markersize=12)
plt.grid(True)
```

```python
## Multiple Plots
## Sample Data

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.figure(figsize=(9, 5))

plt.subplot(1, 2, 1)

plt.plot(x, y1, color='green')

plt.title("Plot 1")
```
